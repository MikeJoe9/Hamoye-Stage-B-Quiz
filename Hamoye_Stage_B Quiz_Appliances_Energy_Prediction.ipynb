```python
df1 = df.drop(columns=['date','lights'])
df1
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
normalised = pd.DataFrame(scaler.fit_transform(df1), columns=df1.columns)
print(normalised)
```

```python
x1 = normalised.drop(columns=['Appliances'])
y1 = normalised['Appliances']
```

```python
from sklearn.model_selection import train_test_split

x1_train, x1_test, y1_train, y1_test = train_test_split(x1, y1,test_size=0.3, random_state=42)
print(x1_train.shape)
print(y1_train.shape)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
model.fit(x1_train, y1_train)
```

```python
y1_pred = model.predict(x1_test)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

mae = mean_absolute_error(y1_test, y1_pred)
print(round(mae,2) )
```

```python
import numpy as np
res_sum_sqrs = np.sum(np.square(y1_test - y1_pred))
print(round(res_sum_sqrs, 2)) 
```

```python
from sklearn.metrics import mean_squared_error
mse = mean_squared_error(y1_test, y1_pred)
rmse =np.sqrt(mean_squared_error(y1_test, y1_pred))
print(round(rmse,3))
```

```python
from sklearn.metrics import r2_score
r2_score = r2_score(y1_test, y1_pred)
round(r2_score, 2)
```

```python
from sklearn.linear_model import Lasso
lasso_reg = Lasso(alpha=0.01)
lasso_reg = lasso_reg.fit(x1_train, y1_train)

from sklearn.linear_model import Ridge
ridge_reg = Ridge(alpha=0.5)
ridge_reg.fit(x1_train, y1_train)
```

```python
def get_weights_df(model, feat, col_name):
    #this function returns the weight of every feature 
    weights = pd.Series(model.coef_, feat.columns).sort_values()
    weights_df = pd.DataFrame(weights).reset_index() 
    weights_df.columns = ['Features', col_name] 
    weights_df[col_name].round(3)
    return weights_df
```

```python
linear_model_weights = get_weights_df(model, x1_train, 'Linear_Model_Weight') 
ridge_weights_df = get_weights_df(ridge_reg, x1_train, 'Ridge_Weight') 
lasso_weights_df = get_weights_df(lasso_reg, x1_train, 'Lasso_weight') 
final_weights = pd.merge(linear_model_weights, ridge_weights_df, on='Features') 
final_weights = pd.merge(final_weights, lasso_weights_df, on='Features')
```
